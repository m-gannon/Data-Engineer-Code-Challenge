# Setup

```python
# Library import
import pandas as pd
import numpy as np
```

# Obtain airline dataset

```python
# URL
url = 'https://data.gov.au/data/dataset/ad89b4ff-541a-4729-b93c-4d2f5682e4c8/resource/809c77d8-fd68-4a2c-806f-c63d64e69842/download/airline_portcountry.csv'

# Create dataframe from csv file at specified URL
df = pd.read_csv(url)
df
```

# 1.	How many passengers flew into and out of Australia for each airline in each month of 2019?

```python
# Filter data for 2019
df2019 = df[(df['Year']==2019)]

# group data by Airline and Month, displaying the sum of Passengers_In and Passengers_Out
print("The number of passengers who flew into and out of Australia in each month of 2019 is displayed in the table below for each airline: ")
group = df2019.groupby(["Airline", "Month_num"])
dfQ1 = group.sum()[[("Passengers_In"),("Passengers_Out")]]
dfQ1
```

# 2.	In the last 6 months which port country was responsible for the most passengers coming into the country?

```python
# Filter data for last 6 moths
df6Mo = df[(df['Month']>=44197)]
df6Mo

# group data by Port country, calculating the sum of Passengers_In 
group = df6Mo.groupby(["Port_Country"])
dfQ2 = group.sum()

# Find the Airline with the maximum sum of Passengers_In

column = dfQ2["Passengers_In"]
max_index = column.idxmax()
max_value = column.max()

print("In the last 6 months, the port country responsible for the most passengers coming into the country is",max_index, "with", max_value, "Passengers")

```

# 3.	What is the 3 month rolling average of freight arriving to Australia for each airline throughout 2018?

```python
# Filter data for 2018
df2018 = df[(df["Year"]==2018)]

# Group by Airline

grouped_dataframe = df2018.groupby("Airline")['Freight_In_(tonnes)']

# calculate 3 month rolling average for Freight_in for each airline
for Airline, data in grouped_dataframe:
    print ("The 3 month rolling average of freight arriving into Austalia on", Airline, "throughout 2018 is displayed in the array below:
           ")
    print ("\n")
    print (data.rolling(window=3).mean()) 
    print ("\n")
    
```
